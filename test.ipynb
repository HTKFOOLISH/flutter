## Folder 4: Farm ManagementManagement
Contains Dart code for a farm-related test.  
**Description**:    
Nhật ký trồng trọt và truy xuất nguồn gốc:
- Nhật ký trồng trọt: có chụp hình
	- giống cây
	- chăm sóc, bón phân,
	- phun thuốc
	- thu hoạch
	- bảo quản
- Truy xuất nguồn gốc: Quét mã QRCode
	- Đặc điểm 
	- Cơ sở sản xuất
	- Thông tin canh tác
